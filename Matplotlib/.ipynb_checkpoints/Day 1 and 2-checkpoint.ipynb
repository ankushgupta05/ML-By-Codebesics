```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(x,y, color='green', linewidth=2, linestyle='dotted')
```
