```python
import pandas as pd
import numpy as np
```

```python
india_weather = pd.DataFrame({
    'city':['mumbai','delhi','banglore'],
    'temperature':[32,45,30],
    'humidity':[80,60,75]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    'city':['patmumbai','usdelhi','usbanglore'],
    'temperature':[322,435,330],
    'humidity':[801,620,725]
})

us_weather
```

# concat()

```python
df = pd.concat([india_weather,us_weather])
df
```

```python
df = pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
df = pd.concat([india_weather,us_weather],keys=['india','us'])
df
```

```python
df.loc['india']
```

```python
temperature_weather = pd.DataFrame({
    'city':['patmumbai','usdelhi','usbanglore'],
    'temperature':[322,435,330],
})

temperature_weather
```

```python
windspeed_weather = pd.DataFrame({
    'city':['patmumbai','usdelhi','usbanglore'],
    'windspeed':[801,620,725]

})

windspeed_weather
```

```python
pd.concat([temperature_weather,windspeed_weather])
```

```python
pd.concat([temperature_weather,windspeed_weather],axis=1)
```
